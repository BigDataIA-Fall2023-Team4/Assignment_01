```python
# importing requirements
from diagrams import Diagram, Edge
from diagrams.custom import Custom
from diagrams.programming.language import Python
```

```python
# Diagram builder
with Diagram("PDF Parsing Process", show=False) as diag:
  # Fetching the required logos through custom locations
    user = Custom("User", "/content/user.png")
    enduser = Custom("User", "/content/user.png")
    streamlit = Custom("Streamlit App", "/content/streamlit logo.png")
    streamlit_dashboard = Custom("Streamlit Dashboard", "/content/streamlit logo.png")
    python = Python("Python Code")
    colab = Custom("Google Colab", "/content/colab_logo.png")
    nougat = Custom("Nougat", "/content/nougat_logo.png")
    pypdf = Custom("Pypdf", "/content/pypdf.png")
    localtunnel = Custom("Local Tunnel", "/content/localtunnel.png")
    spacy = Custom("Spacy", "/content/SpaCy_logo.svg.png")
  # providing the flow of each element in architecture diagram
    user >> Edge(label = "Adding PDF URL and selecting python library") >> streamlit
    streamlit >> Edge(label = "Calling Python Functions") >> python
    python >> Edge(label = "PyPDF library chosen") >> pypdf
    python >> Edge(label = "Nougat library chosen") >> localtunnel
    colab >> Edge(label = "Initializing Nougat API") >> localtunnel
    localtunnel >> Edge(label = "Processing on Nougat API") >> nougat
    pypdf >> Edge(label = "Text summarization") >> spacy
    nougat >> Edge(label = "Text summarization") >> spacy
    spacy >> Edge(label = "View Parsed PDF") >> streamlit_dashboard
    streamlit_dashboard >> Edge(label = "User consumption") >> enduser

diag
```
